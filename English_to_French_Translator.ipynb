```python
from transformers import MarianMTModel, MarianTokenizer

#MarianMTModel for sequence-to-sequence translation
#MarianTokenizer for converting text into tokens (numerical input the model can understand) and back into human-readable text after translation

model_name = 'Helsinki-NLP/opus-mt-en-fr'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
def translate_to_french(en):
  try:
    # Tokenize input text, return_tensors="pt" to return data as PyTorch Tensors, padding ensures and converts all sentences to same length.
    inputs = tokenizer(en, return_tensors="pt", padding=True)     #inputs consists of input_ids i.e., tokens and attention_masks i.e., telling which are words and which are padding
    # Generate translation, **inputs unpacks the input dictionary to keyword arguments i.e., passing all tokens from tokenizer output to model
    translated = model.generate(**inputs)     #stores translated text in token form
    # Decode translation, translated[0] gives the translated text, skip_special_tokens=True removes all special tokens like <pad> etc.
    return tokenizer.decode(translated[0], skip_special_tokens=True)    #returning actual translated text
  except Exception as e:
    return f"Error ocurred as {e}"
```

```python
from gtts import gTTS         #it is module used for text to speech given by google
from IPython.display import Audio         #to play the generated audio
def recite_en(text, lang='en'):              #recite_en() takes argument as english text , converts to audio , which then gets played
    tts = gTTS(text=text, lang=lang)
    tts.save("audio.mp3")
    return Audio("audio.mp3", rate=24000 , autoplay=True)
def recite_fr(text, lang='fr'):              #recite_fr() takes argument as french text , converts to audio , which then gets played
    tts = gTTS(text=text, lang=lang)
    tts.save("audio.mp3")
    return Audio("audio.mp3", rate=24000 , autoplay=True)
```

```python
def flow_en(t):          #flow_en() takes english text argument and prints letters with a slight delay giving effect of words getting typed , also reads the output inhand through recite_en()
  display(recite_en(t)) #display() forces Colab to render the audio player, even inside loops or deeply nested function calls
  print("Bot: ",end='')
  for letter in t:
    print(letter,end='',flush=True) #flush: to avoid storage in buffer and continue
    time.sleep(0.08)
  print()
def flow_fr(t):          #flow_fr() takes french text argument and prints letters with a slight delay giving effect of words getting typed , also reads the output inhand through recite_fr()
  display(recite_fr(t)) #display() forces Colab to render the audio player, even inside loops or deeply nested function calls
  print("Bot: ",end='')
  for letter in t:
    print(letter,end='',flush=True) #flush: to avoid storage in buffer and continue
    time.sleep(0.08)
  print()
```

```python
from google.colab import files      #for writing in file stored in google colab
```

```python
flow_en("Welcome to English-to-French translator, type 'exit' to exit.\n")
time.sleep(1.1)       #to avoid overlapping of 2 speeches
while True:           #input loop
  flow_en("Enter text to translate. Type 'exit' to exit.\n")
  time.sleep(1.1)       #to avoid overlapping of 2 speeches
  p=input("You: ")
  if p.lower()=="exit":     #exit the loop
    flow_en("Thanks for coming!!!")
    break
  else:
    trans=translate_to_french(p)    #main function call
    flow_fr(trans)
    time.sleep(1.1)       #to avoid overlapping of 2 speeches
    flow_en("Do you want to download the translated text?")
    time.sleep(1.1)       #to avoid overlapping of 2 speeches
    if (input()).lower()=="yes":        #download the translated text
      f=open("translated.txt","a")
      f.write("\n\n")
      f.write(trans)
      f.flush()     #to avoid storage in buffer and continue
      f.close()
      files.download("translated.txt")

```
